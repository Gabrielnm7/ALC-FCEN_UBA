```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
años = np.array([1950, 1960, 1970, 1980, 1990, 2000])
habitantes = np.array([17, 20.5, 23.9, 27.9, 32.6, 36.9])

#polinomio de grado 1 a los datos
coeficientes = np.polyfit(años, habitantes, 1)
a, b = coeficientes

f = lambda x: a * x + b

años_ajuste = np.linspace(1950, 2000, 100)
habitantes_ajuste = f(años_ajuste)

# Graficamos
plt.figure(figsize=(8, 6))
plt.scatter(años, habitantes, color='blue', label='Datos originales')
plt.plot(años_ajuste, habitantes_ajuste, color='red', label=f'Ajuste lineal: $f(x) = {a:.2f}x + {b:.2f}$')
plt.xlabel('Año')
plt.ylabel('Población (millones)')
plt.title('Ajuste lineal de la población argentina')
plt.grid(True)
plt.legend()
plt.show()
```
